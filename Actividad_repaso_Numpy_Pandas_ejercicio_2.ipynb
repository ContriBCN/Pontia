```python
df = pd.read_csv('/content/streaming.csv')
print(type(df))
```

```python
df.head(2)
```

```python
# 1. Eliminar las columnas con información de id de sesiones, id de dispositivos, lenguaje, localización, calidad de playback y
#cantidad de interacciones
df_nuevo = df.drop(["Session_ID", "Device_ID", "Languages","Location","Playback_Quality", "Interaction_Events"], axis=1)

df_nuevo.head(5)
```

```python
# 2. Genero preferido por los hombres
df_hombres = df[df["Gender"] == "Male"]
df_hombres

df_hombres.Genre.value_counts() #Este valor muestra una lista de mayor a menor de todos los Genre
```

```python
# 3. Corroborar si los smartphones son el dispositivo en el cual permanecen más tiempo los usuarios mirando streaming
df_dispositivos = df.groupby(["Device_Type"]).agg({"Duration_Watched (minutes)": "mean"})

df_dispositivos.sort_values(by="Duration_Watched (minutes)", ascending=False)



```

```python
# 4. Las edades que suelen pagar más Premium
df_premium = df[df["Subscription_Status"]=="Premium"]

df_premium = df.groupby (["Age"]).agg({"Age":"count"})

df_premium=  df_premium.rename(columns={"Age":"Total"})

df_premium.sort_values(by="Total", ascending=False).head(5)
```

```python
# 5. El video que tiene más tiempo de reproducción

df_maxminutes = df.sort_values(by="Duration_Watched (minutes)", ascending=False).iloc[0]
df_maxminutes
```

```python
# 6. El género que más le gustó a las mujeres (rating)

df.head(5)
df_rating = df[df["Gender"]=="Female"]
df_rating.groupby(["Video_ID"]).agg({"Ratings":"mean"}).sort_values(by="Ratings", ascending=False)

```
