## Array

- Numpy: Una biblioteca en Python para soporte de arrays de varias dimensiones. No es parte del est치ndar de Python, por lo que debe ser instalada (usualmente con pip install numpy).

- Arrays Homog칠neos: A diferencia de las listas de Python, los arrays en Numpy deben contener elementos del mismo tipo, lo que los hace m치s eficientes para c치lculos num칠ricos.

- Dimensiones M칰ltiples: Los arrays pueden ser unidimensionales (vectores), bidimensionales (matrices), o de mayor dimensi칩n para datos m치s complejos.

- Eficiencia: Numpy est치 optimizado para operaciones matem치ticas y manejo de datos num칠ricos, permitiendo operaciones vectorizadas que son m치s r치pidas que las operaciones equivalentes en listas de Python

```python
a_lista = [1, 2, 3]
print()
# Creamos un array unidimensional con los elementos 1, 2, 3
a = np.array([1, 2, 3])

# Creamos un array bidimensional; cada par de par칠ntesis representa una fila
b = np.array([(1, 2, 3), (4, 5, 6)])

print(b)
# Creamos un array tridimensional, pensado como una lista de listas de listas
c = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print(c)
```

```python
# Creando una lista con elementos 1, 2, 3
a2 = [1, 2, 3]

# Imprimiendo el array 'a' creado anteriormente con Numpy
print(f"Array: {a}")
# Imprimiendo la lista 'a2'
print(f"Lista: {a2}")
# Imprimiendo el tipo de 'a' para demostrar que es un array de Numpy
print(type(a))
# Imprimiendo el tipo de 'a2' para demostrar que es una lista de Python
print(type(a2))

```

## Vector

Un vector es un array de **1 fila** x **N columnas** (vector de fila) o **1 columna** x **N filas** (vector de columna).

```python
# Definimos dos vectores con Numpy
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])

# Suma de dos vectores
# Realiza la suma elemento a elemento de los vectores
resultado = vector1 + vector2
print(resultado)

# Producto escalar de dos vectores
# Calcula el producto punto, que es la suma de los productos de sus elementos correspondientes
producto_punto = np.dot(vector1, vector2)
print(producto_punto)
```

## Matriz

Una matriz es un array de **M filas** x **N columnas**.

```python
# Generamos un vector de 5 enteros aleatorios entre 0 y 9
vector = np.random.randint(0, 10, 5)
print(vector)

# Otra opci칩n es generar un vector de 5 n칰meros flotantes aleatorios entre 0.0 y 3.0
vector = np.random.uniform(0.0, 3.0, 5)
vector

# Tambi칠n podemos generar un vector de 5 n칰meros flotantes aleatorios entre 0.0 y 1.0
vector = np.random.random(5)
vector

# Para verificar el tipo de 'vector', podemos descomentar la siguiente l칤nea
print(type(vector))
```

```python
# Creamos una matriz de 5 filas y 3 columnas con enteros aleatorios entre 0 y 9
matriz = np.random.randint(0, 10, (5, 3))
# Imprimimos la matriz para visualizar su contenido
print(matriz)
# Verificamos e imprimimos el tipo de dato de 'matriz'
print(type(matriz))
```

Aclaracion de producto vectorial:

A x B = [(A[1] * B[2] - A[2] * B[1]), (A[2] * B[0] - A[0] * B[2]), (A[0] * B[1] - A[1] * B[0])]

En este caso:

A = [1, 8, 5]

B = [5, 8, 0]

Calculando el producto vectorial:

A x B = [(8 * 0 - 5 * 8), (5 * 1 - 5 * 0), (1 * 8 - 8 * 8)]

A x B = (-40, 5, -32)

Merge/Joins

```python
# Creaci칩n de dos DataFrames a partir de diccionarios
data1 = {'ID': [1, 2, 3, 4],
         'Nombre': ['Alice', 'Bob', 'Charlie', 'David']}
df1 = pd.DataFrame(data1)

data2 = {'ID': [2, 3, 4, 5],
         'Edad': [25, 30, 35, 40]}
df2 = pd.DataFrame(data2)
```

```python
# Inner Merge (Fusi칩n Interna)
# Combina df1 y df2 bas치ndose en la columna 'ID', incluyendo solo las filas que tienen un ID com칰n en ambos DataFrames.
inner_df = pd.merge(df1, df2, on='ID')
# Muestra el resultado de la fusi칩n interna, que incluye solo las filas con IDs 2, 3, y 4 (comunes en ambos).
inner_df
```

```python
# Left Merge (Fusi칩n Izquierda)
# Combina df1 y df2 bas치ndose en 'ID', incluyendo todas las filas de df1 y solo las filas coincidentes de df2, llenando con NaN donde no hay coincidencia.
left_df = pd.merge(df1, df2, on='ID', how='left')
# Muestra el resultado de la fusi칩n izquierda, conservando todos los IDs de df1.
left_df
```

```python
# Right Merge (Fusi칩n Derecha)
# Similar a la fusi칩n izquierda, pero incluyendo todas las filas de df2 y solo las filas coincidentes de df1.
right_df = pd.merge(df1, df2, on='ID', how='right')
# Muestra el resultado de la fusi칩n derecha, conservando todos los IDs de df2.
right_df
```

```python
# Outer Merge (Fusi칩n Externa)
# Combina df1 y df2 bas치ndose en 'ID', incluyendo todas las filas de ambos DataFrames, llenando con NaN donde no hay coincidencia.
outer_df = pd.merge(df1, df2, on='ID', how='outer')
# Muestra el resultado de la fusi칩n externa, que incluye todos los IDs de df1 y df2.
outer_df
```

```python
###########EXPLICACI칍N###################################

def analysis_gender(df, gender_input):
    # Agrupa el DataFrame por g칠nero y calcula el n칰mero 칰nico de IDs de usuario y la suma de compras por g칠nero
    agrupacion_df = df.groupby("Gender").agg({"User ID": "nunique", "Purchased": "sum"}).reset_index()

    # Calcula la tasa de conversi칩n dividiendo el total de compras por el n칰mero 칰nico de usuarios por g칠nero
    agrupacion_df["Conversion"] = agrupacion_df["Purchased"] / agrupacion_df["User ID"]

    # Imprime el resultado de la tasa de conversi칩n por g칠nero
    print(f"---CONVERSION POR GENERO {gender_input}---")
    display(agrupacion_df)  # Utiliza display para una representaci칩n m치s amigable en Jupyter Notebooks

    # Filtra el DataFrame original para incluir solo los registros del g칠nero especificado
    Gender_df = df[df["Gender"] == gender_input]

    # Agrupa el DataFrame filtrado por edad y suma las compras para cada edad
    agrupacion_edad_Gender = Gender_df.groupby("Age").agg({"Purchased": "sum"}).reset_index()

    # Convierte la columna 'Age' a string para facilitar la creaci칩n de gr치ficos
    agrupacion_edad_Gender["Age"] = agrupacion_edad_Gender["Age"].astype(str)

    # Crea un gr치fico de barras mostrando las compras por cada grupo de edad dentro del g칠nero especificado
    agrupacion_edad_Gender.plot(x="Age", y="Purchased", kind="bar", figsize=(10,5))
```

```python
import random  # Importa el m칩dulo random para generar selecciones aleatorias

# Define una lista de ciudades posibles.
ciudades = ["Madrid", "Barcelona", "Sevilla", "Valencia"]

# Inicializa una lista vac칤a para almacenar la ciudad asignada a cada usuario.
lista_ciudades = []

# Inicializa un contador a 0. Este contador se utilizar치 para iterar sobre todos los usuarios.
count = 0

# Mientras el contador sea menor que el n칰mero total de usuarios, ejecuta el bloque de c칩digo.
while (count < len(usuarios)):
    # A침ade una ciudad elegida aleatoriamente de la lista de ciudades a 'lista_ciudades'.
    # 'random.choice(ciudades)' selecciona un elemento al azar de la lista 'ciudades'.
    lista_ciudades.append(random.choice(ciudades))

    # Incrementa el contador en 1 despu칠s de cada iteraci칩n para asegurar que el bucle avance
    # y eventualmente termine cuando se haya asignado una ciudad a cada usuario.
    count = count + 1
```

```python
# Imprime la longitud de la lista de ciudades para verificar cu치ntas ciudades han sido asignadas.
print(len(lista_ciudades))
# Imprime la longitud de la lista de usuarios 칰nicos para asegurar que cada usuario recibir치 una ciudad.
print(len(usuarios))
```

```python
# Combina las listas de usuarios y ciudades en una lista de tuplas, con cada tupla conteniendo un par de 'User ID' y 'lista_ciudades'.
list_of_tuples = list(zip(usuarios, lista_ciudades))

# Muestra las primeras 5 tuplas para verificar los pares de 'User ID' y su ciudad asignada.
print(list_of_tuples[:5])

# Convierte la lista de tuplas en un DataFrame para una manipulaci칩n m치s f치cil.
# Este nuevo DataFrame tiene columnas para 'User ID' y 'lista_ciudades', asociando cada usuario con una ciudad.
df_to_merge = pd.DataFrame(list_of_tuples, columns=['User ID', 'lista_ciudades'])
df_to_merge.head()
```

```python
# Fusiona el DataFrame original 'df_dummies' con 'df_to_merge', usando 'User ID' como clave de fusi칩n.
# Esto a침ade la columna 'lista_ciudades' a 'df_dummies', proporcionando la informaci칩n de la ciudad para cada usuario.
# Se utiliza un 'left join' para asegurar que todos los registros originales en 'df_dummies' se mantengan.
df_final = df_dummies.merge(df_to_merge, how="left", on="User ID")
```
